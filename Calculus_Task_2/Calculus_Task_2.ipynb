#### 1.	Даны три множества a,b и с. Необходимо выполнить все изученные виды бинарных операций над всеми комбинациями множеств.
#### 2.	*Выполнить задание 1 на языке Python

```python
A = {i ** 2 for i in range(5)}
B = {i * 4 for i in range (5)}
C = {i * 8 for i in range(5)}
print(f'A = {A}\nB = {B}\nC = {C}')
```

```python
print('Объединение множеств:')
print(f'A | B = {A} | {B} = {A | B}')
print(f'B | C = {B | C}')
print(f'A | B | C = {A.union(B.union(C))}\n')

print('Пересечение множеств:')
print(f'A & B = {A} & {B} = {A & B}')
print(f'A & B & C = {A.intersection(B.intersection(C))}\n')

print('Разница множеств:')
print(f'A \ B = A \ A & B = {A - (A & B)}')
print(f'B \ A = {B.difference(A)}')
print(f'A \ (B \ C) = {A.difference(B.difference(C))}\n')

print('Симметриченая разница множеств:')
print(f'A ^ B = {A ^ B}')
print(f'A ^ B ^ C = {A.symmetric_difference(B.symmetric_difference(C))}\n')
```

```python
print('Декартово произведение множеств:')
print(f'A x B = {list(it.product(A,B,repeat=1))}\n')

print('Комбинация бинарных операций над множествами:')
print(f'(A - B) x (A & B & C) = {list(it.product(A - B, A & B & C,repeat=1))}')
print(f'B | C ^ B \ A & C = {B | C ^ B - A & C}')
print(f'(B | C) ^ B \ A & C = {(B | C) ^ B - A & C}')
```

#### 4.	*Предложить оптимизацию алгоритма, полученного в задании 3, ускоряющую его сходимость.

```python
# Определить уравнение зависимости для небольших k и экстраполировать результат
# Как альтернатива, можно также применить бинарный поиск почленно в функции.

def lim_calc2():
    time = []
    k_n = []
    cur_prev = []    
    k = 1
    while k < 1000:
        n_cur = calc(k+1)
        n_prev = calc(k)
        cur_prev.append(n_cur - n_prev)       
        n_prev = n_cur
        k_n.append(k)
        t = timeit.timeit("calc(k+1) - calc(k)", setup="from __main__ import calc", number=1, globals=locals())
        time.append(t)
        # print(f'k + 1 = {k+1}, k = {k}')
        k+=1
        n_cur = calc(k)        
        
        if k % 100 ==0:
            print(f'counting k.....{k}')
    
    return k_n, cur_prev, time

k, diff, time = lim_calc2()
```

```python
df = pd.DataFrame(k, columns=['k'])
df['k+1 - k'], df['time'] = diff, time
df.head()
```
