1. Для любого элемента в диапазоне от 0 до 1 включительно сигнум равен 1. (Утверждение ложно. При y = 0, sgn(y) = 0)
 - Отрицание: $\exists y \in [0;1]: sgn(y)\not=1$
2. Для любого натурального числа больше 2, существуют натуральные x,y,z, для которых справедливо равенство $x^n = y^n + z^n$
 - Отрицание: $\exists n \in N > 2: \forall x,y,z \in N: x^n \not= y^n + z^n$
3. Для любого вещественного числа существует вещественное множество X, для которого справедливо X > x 
 - Отрицание: $\exists x \in R \forall X \in R: X < x$
4. Для любого комплексного числа не существует такого комплексного элемента, для которого справедливо, что x > y, или x < y
 - Отрицание: $\exists x \in C \exists y \in C : x < y$ & $x > y$
5. Для любого элемента на промежутке от 0 до $\frac{\pi}{2}$ включиельно, существует $\varepsilon > 0$, при котором справедливо неравенство $sin$ $y < sin(y + \varepsilon)$ (Утверждение ложно)
 - Отрицание: $\exists y \in [0; \frac{\pi}{2}] \forall \varepsilon > 0: \sin y > \sin(y+\varepsilon)$
6. Для любого элемента на промежутке от 0 включительно до $\pi$ невключиельно, существует $\varepsilon > 0$, при котором справедливо $cos$ $y > cos(y + \varepsilon)$ (Утверждение ложно)
 - Отрицание: $\exists y \in [0; \pi] \forall \varepsilon > 0: \cos y < \cos(y+\varepsilon)$
7. Существует число, не принадлежащее к натуральным, целым, рациональным, вещественным и комплексным числам (Утверждение ложно)
 - Отрицание: $\forall x: x \notin \{N,Z,Q,R,C\}$

```python
print('Объединение множеств:')
print(f'A | B = {A} | {B} = {A | B}')
print(f'B | C = {B | C}')
print(f'A | B | C = {A.union(B.union(C))}\n')

print('Пересечение множеств:')
print(f'A & B = {A} & {B} = {A & B}')
print(f'A & B & C = {A.intersection(B.intersection(C))}\n')

print('Разница множеств:')
print(f'A \ B = A \ A & B = {A - (A & B)}')
print(f'B \ A = {B.difference(A)}')
print(f'A \ (B \ C) = {A.difference(B.difference(C))}\n')

print('Симметриченая разница множеств:')
print(f'A ^ B = {A ^ B}')
print(f'A ^ B ^ C = {A.symmetric_difference(B.symmetric_difference(C))}\n')
```
